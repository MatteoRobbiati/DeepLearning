```python
data = sklearn.datasets.load_iris()
data['data'][8]
data['target_names']
```

```python
df = pd.DataFrame({'x_1': list(data['data'].T[0]), 'x_2': list(data['data'].T[1]),
                   'x_3': list(data['data'].T[2]), 'x_4': list(data['data'].T[3]),
                   'label': list(data['target'])})
df
```

```python
sns.pairplot(data=df, hue="label", corner=True, palette='CMRmap_r')
plt.show()
```

```python
#questa la versione da pro 

label = pd.get_dummies(df['label'], prefix='label')
df = pd.concat([df,label], axis=1)
df.drop(['label'], axis=1, inplace=True)
```

```python
# questa era la mia versione da nabbo

setosa_list = []
versicolor_list = []
virginica_list = []

for i in range(len(df)):
    if(df['type'][i]==0):
        setosa_list.append(1)
        versicolor_list.append(0)
        virginica_list.append(0)
    elif(df['type'][i]==1):
        setosa_list.append(0)
        versicolor_list.append(1)
        virginica_list.append(0)
    elif(df['type'][i]==2):
        setosa_list.append(0)
        versicolor_list.append(0)        
        virginica_list.append(1)
        
df['label_setosa']=setosa_list
df['label_versicolor']=versicolor_list
df['label_virginica']=virginica_list
```

```python
df.head(5)
```

```python
# roba da nabbo

setosa_df     = df[df['label']==0]
versicolor_df = df[df['label']==1]
virginica_df  = df[df['label']==2]

lenghts = [len(setosa_df), len(versicolor_df), len(virginica_df)]
cutoff = 40
```

```python
# roba da nabbo

setosa_train = setosa_df[:cutoff]
setosa_test  = setosa_df[cutoff:]

print(len(setosa_train), len(setosa_test))

versicolor_train = versicolor_df[:cutoff]
versicolor_test  = versicolor_df[cutoff:]

virginica_train = virginica_df[:cutoff]
virginica_test  = virginica_df[cutoff:]
```

```python
# pro

train_dataset = df.sample(frac=0.8, random_state=1)
test_dataset  = df.drop(train_dataset.index)        #escludo tutte quelle estratte per comporre il train set
train_dataset.head(5)
```

```python
X_train = train_dataset[['x_1', 'x_2', 'x_3', 'x_4']]
Y_train = train_dataset[['label_0', 'label_1', 'label_2']]

X_test = test_dataset[['x_1', 'x_2', 'x_3', 'x_4']]
Y_test = test_dataset[['label_0', 'label_1', 'label_2']]


Y_train.shape
```

```python
def spawn_NN():
    """ Spawn a DNN, this is my first comment #amazing
    
        Args: 
        opt(keras.optimizer) : your choice about the optimizer
    """
    NN = Sequential()
    NN.add(Dense(64, input_shape = (4,), activation='relu'))
    NN.add(Dense(128, activation='relu'))
    NN.add(Dense(128, activation='relu'))
    NN.add(Dense(128, activation='relu'))
    NN.add(Dense(64, activation='relu'))
    NN.add(Dense(3, activation='softmax'))
    NN.compile(loss=keras.losses.categorical_crossentropy, optimizer='adam', metrics=['accuracy'])
    return NN
```

```python
myNN_ES = spawn_NN()
myNN    = spawn_NN()
hist_ES = myNN_ES.fit(x=X_train, y=Y_train, batch_size=32, epochs=200, validation_split=0.4,
               callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)])
hist = myNN.fit(x=X_train, y=Y_train, batch_size=32, epochs=200, validation_split=0.4)
```

```python
def plot_graph(hist):
    plt.figure(figsize=(15,4))
    plt.subplot(1,2,1)
    plt.plot(hist.history['loss'])
    plt.plot(hist.history['val_loss'])
    plt.subplot(1,2,2)
    plt.plot(hist.history['accuracy'])
    plt.plot(hist.history['val_accuracy'])   
```

```python
plot_graph(hist)
plot_graph(hist_ES)
```
